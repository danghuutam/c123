```python
# Nhập thư viện Numpy
import numpy as np

# Câu 1: Tao array arr có kích thước 3x3 với các giá trị True
arr = np.full((3, 3), True) # Tạo array có kích thước 3x3 và điền giá trị True

# Hiển thị arr
print("arr là:", arr)

# Câu 2: Cho arr_ID = np.array([0 1 2 3 5 6 7 8]) . Tạo array 2 chiều có kích thước 3x3 từ art_ID, và lưu vào arr_2D Trong arr_2D, chuyển cột 1 sang cột 3 và ngược lại
arr_ID = np.array([0, 1, 2, 3, 5, 6, 7, 8]) # Tạo array 1 chiều từ list

# Sử dụng hàm np.reshape để tạo array 2 chiều có kích thước 3x3 từ array 1 chiều
arr_2D = np.reshape(arr_ID, (3, 3))

# Hiển thị arr_2D
print("arr_2D là:", arr_2D)

# Sử dụng hàm np.swapaxes để chuyển cột 1 sang cột 3 và ngược lại
arr_2D = np.swapaxes(arr_2D, 0, 1) # Hoán đổi trục 0 và trục 1

# Hiển thị arr_2D
print("arr_2D sau khi chuyển cột là:", arr_2D)

# Câu 3: Từ arr_2D của câu 2 (sau khi đổi thứ tự cột), chuyển dòng 1 sang dòng 2 và ngược lại
# Sử dụng hàm np.swapaxes để chuyển dòng 1 sang dòng 2 và ngược lại
arr_2D = np.swapaxes(arr_2D, 1, 2) # Hoán đổi trục 1 và trục 2

# Hiển thị arr_2D
print("arr_2D sau khi chuyển dòng là:", arr_2D)

# Câu 4: Từ arr_2D của câu 3, đào ngược các dòng của arr_2D
# Sử dụng hàm np.flip để đảo ngược các dòng của array
arr_2D = np.flip(arr_2D, 0) # Đảo ngược trục 0

# Hiển thị arr_2D
print("arr_2D sau khi đào ngược các dòng là:", arr_2D)

# Câu 5: Từ arr_2D của câu 4, đào ngược các cột của arr_2D
# Sử dụng hàm np.flip để đảo ngược các cột của array
arr_2D = np.flip(arr_2D, 1) # Đảo ngược trục 1

# Hiển thị arr_2D
print("arr_2D sau khi đào ngược các cột là:", arr_2D)

# Câu 6: Cho arr_2D_null = np.array([[1, 2, 3], [np.NaN, 5, 6], [7, np.NaN, 9], [4, 5, 6]]) Kiểm tra trong array có giá trị rỗng không?
arr_2D_null = np.array([[1, 2, 3], [np.NaN, 5, 6], [7, np.NaN, 9], [4, 5, 6]]) # Tạo array 2 chiều có chứa giá trị rỗng

# Sử dụng hàm np.isnan để kiểm tra giá trị rỗng trong array
null_check = np.isnan(arr_2D_null) # Trả về array boolean có cùng kích thước với array ban đầu

# Hiển thị null_check
print("null_check là:", null_check)

# Câu 7: Từ arr_2D_null của câu 6, thay thế giá trị null bằng 0
# Sử dụng hàm np.nan_to_num để thay thế giá trị rỗng bằng 0
arr_2D_null = np.nan_to_num(arr_2D_null) # Trả về array mới có cùng kích thước với array ban đầu

# Hiển thị arr_2D_null
print("arr_2D_null sau khi thay thế giá trị null bằng 0 là:", arr_2D_null)
```
