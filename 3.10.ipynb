```python
# Nhập thư viện pandas
import pandas as pd

# Câu 1: Đọc dữ liệu từ tập tin drinks.csv với index_col là cột đầu tiên của dữ liệu, và lưu vào biến drink
drink = pd.read_csv("drinks.csv", index_col=0)

# Cho biết kiểu dữ liệu, kích thước của drink
print(type(drink)) # <class 'pandas.core.frame.DataFrame'>
print(drink.shape) # (193, 6)

# Hiển thị tên các cột của drink
print(drink.columns) # Index(['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol', 'continent'], dtype='object')

# Xem 5 dòng dữ liệu đầu tiên và cuối cùng của drink
print(drink.head()) # In ra 5 dòng đầu
#              beer_servings  spirit_servings  wine_servings  total_litres_of_pure_alcohol continent
# country                                                                                            
# Afghanistan              0                0              0                           0.0      Asia
# Albania                 89              132             54                           4.9    Europe
# Algeria                 25                0             14                           0.7    Africa
# Andorra                245              138            312                          12.4    Europe
# Angola                 217               57             45                           5.9    Africa
print(drink.tail()) # In ra 5 dòng cuối
#             beer_servings  spirit_servings  wine_servings  total_litres_of_pure_alcohol      continent
# country                                                                                              
# Venezuela             333              100              3                           7.7  South America
# Vietnam               111                2              1                           2.0           Asia
# Yemen                   6                0              0                           0.1           Asia
# Zambia                 32               19              4                           2.5         Africa
# Zimbabwe               64               18              4                           4.7         Africa

# Câu 2: Cho biết số lượng bia tiêu thụ trung bình ở mỗi châu lục
print(drink.groupby("continent")["beer_servings"].mean())
# continent
# Africa            61.471698
# Asia              37.045455
# Europe           193.777778
# North America    145.434783
# Oceania           89.687500
# South America    175.083333
# Name: beer_servings, dtype: float64

# Câu 3: Cho biết thông tin thống kê tổng quát số lượng rượu vang được tiêu thụ ở mỗi châu lục
print(drink.groupby("continent")["wine_servings"].describe())
#              count        mean        std  min    25%    50%     75%    max
# continent                                                                  
# Africa        53.0   16.264151  38.846419  0.0   1.00   2.00   13.00  233.0
# Asia          44.0    9.068182  21.667034  0.0   0.00   1.00    8.00  123.0
# Europe        45.0  142.222222  97.421738  0.0  59.00  128.00  195.00  370.0
# North America 23.0   24.521739  28.266378  1.0   5.00  11.00   34.00  100.0
# Oceania       16.0   35.625000  64.555790  0.0   1.00   8.50   23.25  212.0
# South America 12.0   62.416667  88.620189  1.0   3.00  12.00   98.50  221.0

# Câu 4: Cho biết số lượng các loại bia và rượu tiêu thụ trung bình ở mỗi châu lục
print(drink.groupby("continent")[["beer_servings", "spirit_servings", "wine_servings"]].mean())
#              beer_servings  spirit_servings  wine_servings
# continent                                                 
# Africa            61.471698        16.339623      16.264151
# Asia              37.045455        60.840909       9.068182
# Europe           193.777778       132.555556     142.222222
# North America    145.434783       165.739130      24.521739
# Oceania           89.687500        58.437500      35.625000
# South America    175.083333       114.750000      62.416667

# Câu 5: Cho biết giá trị trung vị cho các loại bia và rượu tiêu thụ ở mỗi châu lục
print(drink.groupby("continent")[["beer_servings", "spirit_servings", "wine_servings"]].median())
#              beer_servings  spirit_servings  wine_servings
# continent                                                 
# Africa                32.0              3.0            2.0
# Asia                  17.5             16.0            1.0
# Europe               219.0            122.0          128.0
# North America        143.0            137.0           11.0
# Oceania               52.5             37.0            8.5
# South America        162.5            108.5           12.0

# Câu 6: Cho biết số lượng rượu mạnh tiêu thụ trung bình, lớn nhất và nhỏ nhất ở mỗi châu lục
print(drink.groupby("continent")["spirit_servings"].agg(["mean", "max", "min"]))
#                    mean  max  min
# continent                        
# Africa         16.339623  152    0
# Asia           60.840909  326    0
# Europe        132.555556  373    0
# North America 165.739130  438   68
# Oceania        58.437500  254    0
# South America 114.750000  302   25

# Câu 7: Sắp xếp dữ liệu tăng dần theo số lượng bia tiêu thụ
drink_sorted = drink.sort_values(by="beer_servings")

# Cho biết 5 quốc gia có lượng tiêu thụ bia nhiều nhất
print(drink_sorted.tail(5))
#             beer_servings  spirit_servings  wine_servings  total_litres_of_pure_alcohol      continent
# country                                                                                              
# Venezuela             333              100              3                           7.7  South America
# Poland                343              215             56                          10.9         Europe
# Namibia               376                3              1                           6.8         Africa
# Czech Republic        361              170            134                          11.8         Europe
# Germany               346              117            175                          11.3         Europe

# Cho biết 5 quốc gia có lượng tiêu thụ bia ít nhất
print(drink_sorted.head(5))
#              beer_servings  spirit_serv
```
