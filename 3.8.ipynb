```python
# Nhập thư viện Numpy
import numpy as np

# Mở tập tin dữ liệu heights.txt và positions.txt
with open("heights.txt", "r") as f:
    heights = f.read().splitlines() # Đọc dữ liệu và tách thành list

with open("positions.txt", "r") as f:
    positions = f.read().splitlines() # Đọc dữ liệu và tách thành list

# Câu 1: a) Tạo numpy array np_positions từ list positions. Xem kiểu dữ liệu của np_positions b) Tạo numpy array np_heights từ list heights. Xem kiểu dữ liệu của np_heights
np_positions = np.array(positions) # Chuyển list thành array
np_heights = np.array(heights, dtype=float) # Chuyển list thành array và định dạng kiểu dữ liệu là float

# Hiển thị kiểu dữ liệu của np_positions và np_heights
print("Kiểu dữ liệu của np_positions là:", np_positions.dtype)
print("Kiểu dữ liệu của np_heights là:", np_heights.dtype)

# Câu 2: Tính chiều cao trung bình của các GK
gk_heights = np_heights[np_positions == 'GK'] # Lọc các phần tử có vị trí là GK
gk_mean = np.mean(gk_heights) # Tính giá trị trung bình của array

# Hiển thị gk_mean
print("Chiều cao trung bình của các GK là:", gk_mean)

# Câu 3: Tính chiều cao trung bình của những vị trí khác (Không phải là GK)
other_heights = np_heights[np_positions != 'GK'] # Lọc các phần tử có vị trí khác GK
other_mean = np.mean(other_heights) # Tính giá trị trung bình của array

# Hiển thị other_mean
print("Chiều cao trung bình của những vị trí khác là:", other_mean)

# Câu 4: Tạo mảng dữ liệu có cấu trúc tự định nghĩa players gồm 'position' kiểu văn bản (US) và 'height' kiểu 'float'
# Sử dụng hàm np.dtype để định nghĩa kiểu dữ liệu có cấu trúc
dt = np.dtype([('position', 'U2'), ('height', 'float')]) # U2 là kiểu văn bản có độ dài tối đa là 2 ký tự

# Sử dụng hàm np.array để tạo array có cấu trúc từ hai array np_positions và np_heights
players = np.array(list(zip(np_positions, np_heights)), dtype=dt)

# Hiển thị players
print("players là:", players)

# Câu 5: Sắp mảng players theo height, cho biết vị trí có chiều cao cao nhất và chiều cao thấp nhất
# Sử dụng hàm np.sort để sắp xếp array theo trường height
players_sorted = np.sort(players, order='height') # Sắp xếp theo thứ tự tăng dần

# Hiển thị players_sorted
print("players_sorted là:", players_sorted)

# Lấy phần tử đầu tiên và cuối cùng của array
min_height = players_sorted[0] # Phần tử có chiều cao thấp nhất
max_height = players_sorted[-1] # Phần tử có chiều cao cao nhất

# Hiển thị min_height và max_height
print("Vị trí có chiều cao thấp nhất là:", min_height['position'])
print("Chiều cao thấp nhất là:", min_height['height'])
print("Vị trí có chiều cao cao nhất là:", max_height['position'])
print("Chiều cao cao nhất là:", max_height['height'])
```
