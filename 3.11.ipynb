```python
# Import thư viện pandas
import pandas as pd

# Câu 1a: Đọc file stocks1.csv và xem thông tin
stocks1 = pd.read_csv('stocks1.csv')
print(stocks1.head(5)) # Hiển thị 5 dòng đầu
print(stocks1.tail(5)) # Hiển thị 5 dòng cuối
print(stocks1.dtypes) # Hiển thị kiểu dữ liệu của các cột
print(stocks1.info()) # Hiển thị thông tin tổng quan

# Câu 1b: Đọc file stocks2.csv và xem thông tin
stocks2 = pd.read_csv('stocks2.csv')
print(stocks2.head(5))
print(stocks2.tail(5))
print(stocks2.dtypes)
print(stocks2.info())

# Câu 1c: Đọc file companies.csv và xem thông tin
companies = pd.read_csv('companies.csv')
print(companies)
print(companies.dtypes)
print(companies.info())

# Câu 2: Kiểm tra và xử lý dữ liệu Null trong stocks1
print(stocks1.isnull().sum()) # Kiểm tra số lượng dữ liệu Null trong mỗi cột
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max())) # Thay Null cột high bằng max của mã chứng khoán
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min())) # Thay Null cột low bằng min của mã chứng khoán
print(stocks1.isnull().sum()) # Kiểm tra lại số lượng dữ liệu Null sau khi xử lý

# Câu 3: Gộp stocks1 và stocks2 theo dòng
stocks = pd.concat([stocks1, stocks2], axis=0) # axis=0 nghĩa là gộp theo dòng
print(stocks.tail(15)) # Hiển thị 15 dòng cuối của stocks

# Câu 4: Gộp stocks và companies
stocks_companies = pd.merge(stocks, companies, on='symbol', how='left') # Gộp theo cột symbol, how='left' nghĩa là giữ nguyên dữ liệu của stocks
print(stocks_companies.head(5)) # Hiển thị 5 dòng đầu của stocks_companies

# Câu 5: Tính giá trung bình và volume trung bình của mỗi công ty
avg_price_volume = stocks_companies.groupby('name').agg({'open':'mean', 'high':'mean', 'low':'mean', 'close':'mean', 'volume':'mean'}) # Nhóm theo name và tính mean của các cột giá và volume
print(avg_price_volume) # Hiển thị kết quả

# Câu 6: Tính giá đóng cửa trung bình, lớn nhất và nhỏ nhất của mỗi công ty
close_stats = stocks_companies.groupby('name')['close'].agg(['mean', 'max', 'min']) # Nhóm theo name và tính mean, max, min của cột close
print(close_stats) # Hiển thị kết quả

# Câu 7: Tạo cột parsed_time và xem kiểu dữ liệu
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date']) # Đổi thời gian sang định dạng DateTime
print(stocks_companies['parsed_time'].dtype) # Hiển thị kiểu dữ liệu của cột parsed_time
print(stocks_companies.head(5)) # Hiển thị 5 dòng đầu của stocks_companies

# Câu 8: Tạo cột result và gán giá trị
stocks_companies['result'] = stocks_companies.apply(lambda x: 'up' if x['close'] > x['open'] else 'down', axis=1) # Dùng hàm apply để gán giá trị 'up' hoặc 'down' cho mỗi dòng dựa trên điều kiện
print(stocks_companies) # Hiển thị kết quả
```
